```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv("Salary_Data.csv")
df
```

```python
x= df['YearsExperience']
y=df['Salary']
```

```python
plt.scatter(x,y , color='r')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Experience and salary')
plt.show()
```

```python
sns.distplot(x, bins = 10, kde = True)
plt.title('Before Transformation')
plt.show
import warnings
warnings.filterwarnings('ignore')
```

```python
# log Transformation 
```

```python
sns.displot(np.log(x), bins = 10, kde = True)
plt.title('After log Transformation')
plt.show()
```

```python
# SQRT Transformation 
```

```python
sns.displot(np.sqrt(x), bins = 10, kde = True)
plt.title('After SQRT Transformation')
plt.show()
```

```python
stats.pearsonr(x,y)
```

```python
np.corrcoef(x,y)
```

```python
xx= df[['YearsExperience']]
yy=df['Salary']
```

```python
model = LinearRegression()
```

```python
model.fit(xx,yy)
```

```python
yp = model.predict(xx)
yp
```

```python
mse = mean_squared_error(yy, yp)
print('Mean Squared Error:',mse)
```

```python
plt.scatter(x,y, color='r')
plt.plot(x,yp,color='g')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Regression Line')
plt.show()
```
