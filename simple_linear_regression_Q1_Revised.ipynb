```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('delivery_time.csv')
del_time = df[["Delivery Time"]]
sort_time = df["Sorting Time"]
```

```python
print(df)
```

```python
print(df.describe())
```

```python
x=del_time
y=sort_time
```

```python
plt.scatter(x,y)
plt.xlabel('Delivery Tine')
plt.ylabel('Sorting Time')
plt.title('Delivery Time and Sorting Time')
plt.show()
```

```python
sns.distplot(x, bins = 10, kde = True)
plt.title('Before Transformation')
plt.show
import warnings
warnings.filterwarnings('ignore')
```

```python
# log transformation
```

```python
sns.displot(np.log(x), bins = 10, kde = True)
plt.title('After log Transformation')
plt.show()
```

```python
# SQRT Transformation 
```

```python
sns.displot(np.sqrt(x), bins = 10, kde = True)
plt.title('After SQRT Transformation')
plt.show()
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
yp = model.predict(x)
yp
```

```python
mse = mean_squared_error(y, yp)
print('Mean Squared Error:',mse)
```

```python
plt.scatter(x,y, color='r')
plt.plot(x,yp,color='g')
plt.xlabel('Sorting Time')
plt.ylabel('Delivery Time')
plt.title('Regression Line')
plt.show()
```
